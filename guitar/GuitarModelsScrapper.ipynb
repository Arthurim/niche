```python
brands = pd.read_csv(path+"guitar_brands.csv")["Brand"].tolist()
nb_brands=len(brands)
for i in range(nb_brands):
    brand = brands[i]
    print("Processing:"+brand+" - " + str(round(i/nb_brands*100,2))+"%")
    try:    
        url = "https://www.sweetwater.com/c590--"+brand+"--Solidbody_Guitars?params=eyJjb21wYXJlIjpbXX0"
        print("    ",url)
        proxy = next(proxy_pool)
        r = requests.get(url,headers=random.choice(headers_list),proxies={"http": proxy, "https": proxy})
        print(r)
        soup = bs4.BeautifulSoup(r.text, 'html.parser')
        nb_pages = int(soup.findAll("strong", {"class":"page__counter"})[0].text[-1])
        models = []
        
        for page in range(1,nb_pages+1):
            post_url = "https://www.sweetwater.com/c590--"+brand+"--Solidbody_Guitars?pn="+str(page)+"&params=eyJjb21wYXJlIjpbXX0"
            attempts=0
            proxy_success = False
            while attempts < 5:
                print("Proxy attempt #",attempts)
                try:
                    proxy = next(proxy_pool)
                    post_r = requests.get(post_url,headers=random.choice(headers_list),proxies={"http": proxy, "https": proxy})
                    post_soup = bs4.BeautifulSoup(post_r.text, 'html.parser')
                    all_posts = post_soup.findAll("div",{"class":"products"})[0]
                    for post in all_posts.findAll("h2",{"class":"product-card__name"}):
                        #print(post.text)
                        model = post.text.lower()
                        model = model.split(" - ")[0]
                        model = model.replace(brand, "")
                        model = model.replace("\n ","")
                        models.append(model)
                    proxy_success = True
                    break
                except:
                    print("Proxy failed...")
                    attempts += 1      

            if proxy_success:
                post_r.close()
            print("  "+str(round(page/nb_pages*100,2))+"% done")

        models=list(set(models[:-1]))
        pd.DataFrame(models).to_csv(path+"models_"+brand+".csv",index=False)
        r.close()
    except Exception as error:
        print("->Failed.")
        print(error)
```
